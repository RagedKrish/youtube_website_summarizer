```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('House Price India.csv')
df=df.iloc[:,2:]
df.drop(df.columns[13:18], axis=1, inplace=True)
df.drop(df.columns[5:7], axis=1, inplace=True)
df.drop(df.columns[3], axis=1, inplace=True)
df.drop(df.columns[7], axis=1, inplace=True)
df.drop(df.columns[8], axis=1, inplace=True)
df.drop(df.columns[8], axis=1, inplace=True)
df.drop(df.columns[8], axis=1, inplace=True)
df.drop(df.columns[6], axis=1, inplace=True)
df['number of bathrooms'] = df['number of bathrooms'].astype(int)
#df['number of floors'] = df['number of floors'].astype(int)
df.dropna(inplace=True)
df['Price']=df['Price']*10
df
```

```python
def map_grade(x):
    if x <= 5:
        return "Poor"
    elif x <= 7:
        return "Average"
    else:
        return "Very Good"

df['Grade_Category'] = df['grade of the house'].apply(map_grade)
def map_condition(x):
    if x <= 2:
        return "Bad"
    elif x == 3:
        return "Average"
    else:
        return "Good"

df['Condition_Category'] = df['condition of the house'].apply(map_condition)
def map_floors(x):
    if x <= 1.0:
        return "Single"
    else:
        return "Multi"

df['Floor_Type'] = df['number of floors'].apply(map_floors)
df
```

```python
df = pd.get_dummies(df, columns=['Grade_Category', 'Condition_Category', 'Floor_Type'], drop_first=True)
#df.drop(['number of bathrooms', 'condition of the house', 'grade of the house'], axis=1, inplace=True)

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df['space_per_bedroom'] = df.apply(
    lambda row: row['living area'] / row['number of bedrooms'] if row['number of bedrooms'] != 0 else 0,
    axis=1
)

df
```

```python
y=df['Price']
y
```

```python
X = df.drop('Price', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
lrm=LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

X_test_scaled
```

```python

lrm.fit(X_train_scaled, y_train)

```

```python


# Predict and evaluate
y_pred = lrm.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

```python


# Define input values
input_data = {
    'number of bedrooms': 1,
    'number of bathrooms': 2,
    'living area': 1800,
    'number of floors': 1,
    'condition of the house': 4,
    'grade of the house': 8,
    'Built Year': 2010
}

# Feature engineering
input_data['Grade_Category_Poor'] = 1 if input_data['grade of the house'] <= 5 else 0
input_data['Grade_Category_Very Good'] = 1 if input_data['grade of the house'] >= 8 else 0
input_data['Condition_Category_Bad'] = 1 if input_data['condition of the house'] <= 2 else 0
input_data['Condition_Category_Good'] = 1 if input_data['condition of the house'] >= 4 else 0
input_data['Floor_Type_Single'] = 1 if input_data['number of floors'] == 1 else 0


# Final input DataFrame in correct order
input_df = pd.DataFrame([input_data])[[
    'number of bedrooms',
    'number of bathrooms',
    'living area',
    'number of floors',
    'condition of the house',
    'grade of the house',
    'Built Year',
    'Grade_Category_Poor',
    'Grade_Category_Very Good',
    'Condition_Category_Bad',
    'Condition_Category_Good',
    'Floor_Type_Single',
]]

# Scale the input and convert back to DataFrame
scaled_array = scaler.transform(input_df)
scaled_df = pd.DataFrame(scaled_array, columns=input_df.columns)

# Predict
predicted_price = lrm.predict(scaled_df)[0]

print(f"Predicted Price: ₹{round(predicted_price, 2)}")
```
